```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import re
import nltk
from sklearn.datasets import load_files
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
nltk.download('stopwords')
import pickle
from nltk.corpus import stopwords
Temp_Sent=pd.read_csv('twitter.csv')
Temp_Sent
```

```python
Temp_Sent = Temp_Sent.sample(frac = 1)
Temp_Sent
```

```python
Sent = Temp_Sent[100000:200000]
Sent
```

```python
def cleanText (text):    
    import re
    import string
    document = re.sub(r'\d+', '', text)
    document=document.lower()
    document = re.sub(r'[^\w\s]','',document)
    document=document.strip()
    stop_words = set(stopwords.words('english'))
    #tokinized_document= word_tokenize(document)
    #final = [i for i in tokinized_document if not i in stop_words]

    return document
```

```python
document=''
stop = stopwords.words('english')
for index in Sent.index:
    document=cleanText(Sent.loc[index,'text']) 
    Sent.loc[index,'text']=document
Sent['text'] = Sent['text'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))

Sent
```

```python
X_Sent=Sent['text']
Y_Sent=Sent['target']
```

```python
from sklearn import linear_model  
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X_Sent, Y_Sent, test_size=0.25, random_state=1)
```

```python
cv=CountVectorizer()
X_train_Vec1= cv.fit_transform(X_train)
X_train_Vec2= cv.transform(X_test)
a=X_train_Vec1.toarray()
#len(a[0])
a[0]
```

```python
from sklearn.naive_bayes import MultinomialNB
mnb=MultinomialNB()
y_train=y_train.astype('int')
```

```python
mnb.fit(X_train_Vec1,y_train)
```

```python
predcv=mnb.predict(X_train_Vec2)
predcv
```
