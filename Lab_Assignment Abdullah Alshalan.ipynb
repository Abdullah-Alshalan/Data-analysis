```python
# Create a data frame and call it Nasdaq 
# and load all data from nasdaq dataset and display the results.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
LE=pd.read_csv('nasdacq.csv')
LE
```

```python
# Drop all columns rather than date, open, and close
```

```python
LE.drop('Low', axis='columns', inplace=True)
LE.drop('Volume', axis='columns', inplace=True)
LE.drop('High', axis='columns', inplace=True)
LE.drop('Adjusted Close', axis='columns', inplace=True)
LE
```

```python
# Calculate statistical parameters of  ``open``  and `` close`` columns
```

```python
A=LE.groupby('Open').mean()
A
```

```python
C=LE.groupby('Open').max()
C.drop('Date', axis='columns', inplace=True)
C
```

```python
B=LE.groupby('Close').mean()
B
```

```python
D=LE.groupby('Close').max()
D.drop('Date', axis='columns', inplace=True)
D
```

```python
# Remove all null values in ``open`` column and replace it with the mean of the column
```

```python
# be aware the Strings is case sensitive
LE["Open"]=LE["Open"].fillna(np.nanmean(LE["Open"].values))
LE
```

```python
# Add new column and call it `` Diff_Open_close`` 
# and is should contains the difference between open and close price
```

```python
LE['Diff_Open_close']=LE['Open']-LE['Close']
LE
```

```python
# Use scatter diagram to show the relationship between, open, and close
```

```python
plt.scatter(LE['Open'].values, LE['Close'].values)
plt.show()
# Open in x axis
# Close in y axis
```

```python
# Implement a linear regression model to predict the close value depending on open value
# Show the goodness of your results using R2
```

```python
from sklearn.model_selection import train_test_split
x1=LE['Open'].to_frame()
x2=LE['Close']
X_train, X_test, y_train, y_test = train_test_split(x1, x2, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df
```

```python
print('Intercept:', regressor.intercept_)
print('Coefficients:', regressor.coef_)
print('Variance score: {}'.format(regressor.score(x1, x2)))
```

```python
reg = regressor
r2 = reg.score(x1,x2)
print('Intercept:', reg.intercept_)
print('Coefficients:', reg.coef_)
print('Variance score: {}'.format(reg.score(r2)))
```
